```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
home = pd.read_csv("home_stats.csv")
away = pd.read_csv("away_stats.csv")
```

```python
home.head(2)
```

```python
away.head(2)
```

```python
cols = ['del1','del2','Team','MP','Win','Draw','Loss','GF','GA','GD','Points','del3','PPG','CS','BTTS','FTS','del4','del5','1.5+','2.5+','AGPM']
```

```python
len(cols)
```

```python
home.columns = cols
```

```python
home.head()
```

```python
away.columns
```

```python
away.columns = cols
```

```python
away.head()
```

```python
home.drop(['del1','del2','del3','del4','del5'], axis=1, inplace= True)
```

```python
home.head(2)
```

```python
away.drop(['del1','del2','del3','del4','del5'], axis=1, inplace= True)
```

```python
away.head(2)
```

```python
home['Home/Away'] = 'Home'
away['Home/Away'] = 'Away'
```

```python
home.head(2)
```

```python
away.head(2)
```

```python
stats = pd.concat([home,away], axis=0, ignore_index=True)
stats
```

```python
stats.Team = stats.Team.str[0:3].str.upper()
stats.Team
```

```python
stats.Team.iloc[[0,20]] = 'MCI'
stats.Team.iloc[[2,22]] = 'MUN'
stats.Team.iloc[[10,32]] = 'WHU'
stats.Team.iloc[[14,27]] = 'BOU'
stats.Team.iloc[[19,36]] = 'WBA'
stats.Team.iloc[[10,32]] = 'WHU'
```

```python
stats.Team
```

```python
stats.sample(5)
```

```python
stats.replace(regex=True,inplace=True,to_replace=r'%',value=r'')
```

```python
stats.head()
```

```python
stats.info()
```

```python
stats.CS = pd.to_numeric(stats.CS)
stats.BTTS = pd.to_numeric(stats.BTTS)
stats.FTS = pd.to_numeric(stats.FTS)
stats['1.5+'] = pd.to_numeric(stats['1.5+'])
stats['2.5+'] = pd.to_numeric(stats['2.5+'])
```
